```python
import matplotlib.pyplot as plt

def visualize_sales(data):

    products = list(data.keys())
    sales = list(data.values())

    plt.bar(products, sales, color='skyblue')

    plt.xlabel('Products')
    plt.ylabel('Sales Quantity')
    plt.title('Supermarket Sales')

    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

sales_data = {
    'Apple': 50,
    'Banana': 80,
    'Orange': 60,
    'Mango': 45,
    'Grapes': 70
}

visualize_sales(sales_data)
```

```python
import matplotlib.pyplot as plt

def create_pie_chart(products, sales_percentages):
    
    plt.figure(figsize=(8, 8))
    plt.pie(sales_percentages, labels=products, autopct='%1.1f%%', startangle=140)

    plt.title('Supermarket Sales Distribution')

    plt.axis('equal')

    plt.show()

products = ['Apple', 'Banana', 'Orange', 'Mango', 'Grapes']
sales_percentages = [20, 30, 15, 25, 10]

create_pie_chart(products, sales_percentages)
```

```python
import matplotlib.pyplot as plt

def create_line_chart(time_periods, sales):
   
    plt.plot(time_periods, sales, marker='o', linestyle='-')

    plt.xlabel('Time Period')
    plt.ylabel('Sales Quantity')
    plt.title('Supermarket Sales Trend')

    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(True)
    plt.show()

time_periods = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 120, 90, 110, 130, 100]

create_line_chart(time_periods, sales)
```

```python
import matplotlib.pyplot as plt

def create_box_plot(data):
    
    fig, ax = plt.subplots()


    ax.boxplot(data)

    ax.set_xlabel('Data')
    ax.set_ylabel('Values')
    ax.set_title('Box Plot')

    plt.show()

data = [10, 20, 30, 40, 50, 60, 70, 80, 90]

create_box_plot(data)
```

```python
import matplotlib.pyplot as plt

def create_scatter_plot(x_data, y_data):

    fig, ax = plt.subplots()


    ax.scatter(x_data, y_data)

    ax.set_xlabel('X-axis Label')
    ax.set_ylabel('Y-axis Label')
    ax.set_title('Scatter Plot')

    plt.show()

x_data = [1, 2, 3, 4, 5]
y_data = [10, 15, 20, 25, 30]

create_scatter_plot(x_data, y_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')

from scipy.stats import gaussian_kde
kde = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)
plt.plot(x_vals, kde(x_vals), color='red', label='KDE')

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Density Plot with KDE')
plt.legend()

plt.show()
```
